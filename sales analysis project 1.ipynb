```python
sales_table = sales.groupby("Product")[["Total_Sales"]].sum().sort_values("Total_Sales", ascending=False)
```

```python
# Count of each payment method
payment_counts = sales["Payment_Method"].value_counts()

# Plot pie chart
payment_counts.plot(kind="pie", figsize=(6,6), autopct="%1.1f%%", colors=["blue", "orange", "red","green"])

plt.title("Payment Method Distribution")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

```python
# Ensure the "Date" column is in datetime format
sales["Date"] = pd.to_datetime(sales["Date"])

# Extract Year and Month
sales["Year"] = sales["Date"].dt.year
sales["Month"] = sales["Date"].dt.month
```

```python
# Aggregate total sales by month
monthly_sales = sales.groupby(["Year", "Month"])["Total_Sales"].sum().reset_index()

# Convert Year & Month into a readable format
monthly_sales["Month_Year"] = monthly_sales["Year"].astype(str) + "-" + monthly_sales["Month"].astype(str)
```

```python
monthly_sales
```

```python

plt.figure(figsize=(80,10))
plt.plot(monthly_sales["Month_Year"], monthly_sales["Total_Sales"], marker="o", linestyle="-", color="b")

plt.xlabel("Month-Year")
plt.ylabel("Total Sales")
plt.title("Sales Trend Over Time")

plt.show()
```
